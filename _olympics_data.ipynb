### Lets create Seperate columns using if else statements for number of medals


```python
data['Gold'] = [1 if medal == 'Gold' else 0 for medal in data['Medal']]
data['Silver'] = [1 if medal == 'Silver' else 0 for medal in data['Medal']]
data['Bronze'] = [1 if medal =='Bronze' else 0 for medal in data['Medal']]
```

```python
data.head()
```

```python
country_wise_medals = data.groupby(['Team'])[['Gold', 'Silver', 'Bronze']].sum()
country_wise_medals
```

```python
country_wise_medals['total']=country_wise_medals['Gold']+country_wise_medals['Silver']+country_wise_medals['Bronze']
country_wise_medals['total']
```

```python
country_wise_medals.sort_values(by='total',ascending=False)
```

```python
country_wise_medals.nlargest(2,'total')
```

```python
country_wise_medals['total'].max()
```
